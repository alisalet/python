```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,label='sin(x)',color='#40a6e6')
plt.plot(x,z,label='cos(x)',color='#f060c9')
plt.title('Графики sin(x) и cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

```python
x=np.random.normal(0,1,1000)

plt.hist(x,bins=20,color='#40a6e6',edgecolor='#3c7ee8')
plt.title('Гистограмма нормального распределения')
plt.xlabel('Значения')
plt.ylabel('Данные')
plt.show()
```
